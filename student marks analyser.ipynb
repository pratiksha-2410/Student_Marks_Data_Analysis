# Import Liabraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Dataset

```python
df=pd.read_csv('student_marks.csv')
print(df)
```

```python
df.columns.values[0]="Student_Name"
df.columns.values[3]="Maths"
print(df)
```

# Data Inspection

```python
print(df.shape)
print(df.size)
print(df.isna().sum())


```

```python
print(df.info())
```

# Data Cleaning

```python
df.fillna(df.mean(numeric_only=True),inplace=True)
print(df)

```

# Total & Average Marks

```python
subjects=['Maths','Physics','Chemistry','English','Biology', 'Economics','History','Civics']
df["total"]=df[subjects].sum(axis=1)
df["average"]=df["total"]/len(subjects)
print(df[["total","average"]])
```

# Ranking

```python
df["rank"] = df["total"].rank(ascending=False,method='dense')
df.sort_values(by="rank",inplace=True)
print(df["rank"])
```

# Subjectwise Analysis

```python
subjects_analysis = pd.DataFrame({
    "Mean":df[subjects].mean(),
    "Median":df[subjects].median(),
    "max":df[subjects].max(),
    "min":df[subjects].min()
})
print(subjects_analysis)
```

# Percentile Of Each

```python
df["percentage"]=(df["total"]/(len(subjects)*100))*100
print(df.sort_values(by="percentage", ascending=False))
```

# Grade Assignning

```python
def grade(percentage):
    if percentage >= 90:
        return 'A'
    elif percentage >= 80:
        return 'B'
    elif percentage >= 70:
        return 'C'
    elif percentage >= 60:
        return 'D'
    elif percentage >= 50:
        return 'F'
    else:
        return 'FAIL'
    
print(df["percentage"].apply(grade))
```

# Top & Weak Student Analysis


```python
#top 3 students based on percentile
top_student = df.sort_values(by='percentage',ascending=False)[['Student_Name','percentage']].head(3)
print(top_student)
```

```python
#Weak Students
weak_students = df[df["percentage"]<50][['Student_Name','percentage']]
print(weak_students)
```

```python
#Topper rank 1 student details
topper = df.loc[df["percentage"].idxmax()]
print(topper)
```

```python
#weakest student details
weakest_student =df.loc[df["percentage"].idxmin()]
print(weakest_student)
```

```python
#performance visulaization
df["performance"]=pd.cut(df["percentage"],
                         bins=[0,50,60,70,80,90,100],
                         labels=['Fail','Need Improvement','Average','Good','Very Good','Excellent'])
print(df["performance"])
```
